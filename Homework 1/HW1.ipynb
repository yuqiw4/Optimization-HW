### part (a)
$f(x) = |x|$ is convex.
  
Proof:  
Let $a, b\in\mathbb{R}$.  
Let $t\in (0,1)$   
$f((1-t)a + tb)$  
$= |(1-t)a + tb|$  
$\leq |(1-t)a| + |tb|$  
$= (1-t)|a| + t|b|$  
$= (1-t)f(a) + tf(b)$  
So, by definition, $f(x) = |x|$ is convex.

```python
import numpy as np # Namespace is np
import matplotlib.pyplot as plt # Namespace is plt

t = np.linspace(0,1,100)
x = np.linspace(-5,5,1000)
a = -1
b = 3
l = (1-t)*a + t*b

plt.plot(x, np.abs(x))
plt.plot(l, (1-t)*np.abs(a) + t*np.abs(b))
plt.xlabel('x')
plt.ylabel('f(x) = abs(x)')
plt.title('Convex, but not strictly convex')
plt.show()
```

\textbf{Minimum value: 0  
minimizing solution: 0}
  
proof: $|x|\geq0$ for all $x\in\mathbb{R}$, so by the definition of minimum value, $0$ is the minimum value for $f(x)=|x|$.
Then we get 0 is the minimizing solution.

** $f(x) = |x|$ is a Lipschitz function. **  
  
proof: $f(x)$ is continuous and piecewise differentiable on $\mathbb{R}$, since   

$f'(x)$ =
$  \begin{cases}
    -1, & \text{for } x < 0 \\
    1, & \text{for } x > 0 
  \end{cases}  $  
so the program  
$$
\max_{x\in\mathbb{R}}\: |f'(x)| \text{ subject to } f'(x) \text{ existing},
$$  


has maximum value $C = 1 < \infty$. then $f$ is Lipschitz with Lipschitz constant $1$.

### part (d)

Proof:  
We know that $f(x) = |x|$ is convex.  
By the Theorem (Convexity Preservation under Affine Precomposition),  
Let $a = 1, b = -\mu$, so set $Y = \{y\in\mathbb{R}: y - \mu\in\mathbb{R}\}$ is convex.  
And $g:Y\rightarrow\mathbb{R}$ defined by $g(y) = f(y-\mu)$ for all $y\in Y$ is convex on Y.  
Therefore, $f(x) = |x - \mu|$ is convex.

```python
t = np.linspace(0,1,100)
x = np.linspace(-5,5,1000)
a = -1
b = 3
l = (1-t)*a + t*b
mu = 0.5


plt.plot(x, np.abs(x-mu))
plt.plot(l, (1-t)*np.abs(a-mu) + t*np.abs(b-mu))
plt.xlabel('x')
plt.ylabel('f(x) = abs(x-mu)')
plt.title('Convex')
plt.show()
```

**Minimum value: $0$  
minimizing solution: $\mu$  **

  
proof: we know that $|y|\geq0$ for all $y\in\mathbb{R}$, so by the definition of minimum value, $0$ is the minimum value for $f(y)=|y|$.  
Let $y=x-\mu=0$, then we get $\mu$ is the minimizing solution.

** $f(x) = |x - \mu|$ is a Lipschitz function. **  
  
proof: $f(x)$ is continuous and piecewise differentiable on $\mathbb{R}$, since   

$f'(x)$ =
  \begin{cases}

$$

### part (f)
$f(x) = \sum_{i=1}^n |x_i-x|$ where $x_1,...,x_n\in\mathbb{R}$ are fixed values is convex.  

Proof:  
We know that $f(x) = |x|$ is convex.  
By the Theorem (Convexity Preservation under Affine Precomposition),  
Let $a = -1, b = x_i\in\mathbb{R}$, so set $Y = \{y\in\mathbb{R}: -y + x_i\in\mathbb{R}\}$ is convex.  
And $g:Y\rightarrow\mathbb{R}$ defined by $g(y) = f(-y + x_i)$ for all $y\in Y$ is convex on Y.  
Therefore, $f(x) = |x_i - x|$ is convex.  
Then by the Group Question No.3, sum of multiple convex functions is convex.  
Finally, $f(x) = \sum_{i=1}^n |x_i-x|$ where $x_1,...,x_n\in\mathbb{R}$ are fixed values is convex.  


**Minimum value: Plug the median value of ${x_1,x_2,...,x_n}$ into the function $f(x)$ as x.  
minimizing solution: the median value of ${x_1,x_2,...,x_n}$  **

proof: As $x$ equal to the median value of ${x_1,x_2,...,x_n}$, after you remove the absolute value brackets, there will be the same number of $+x$ as the number of $-x$. To the left of the median value, the slope will be negative, so the value of $f(x)$ will increase as $x$ decreases. To the right of the median value, the slope will be positive, so the value of $f(x)$ will increase as $x$ increases. Therefore, we will have a minimizing solution of $f(x)$ as $x$ is the median value of ${x_1,x_2,...,x_n}$.

** $f(x) = \sum_{i=1}^n |x_i-x|$ where $x_1,...,x_n\in\mathbb{R}$ are fixed values is a Lipschitz function. **  
  
proof: $f(x)$ is continuous and piecewise differentiable on $\mathbb{R}$, and the program  
$$
\max_{x\in\mathbb{R}}\: |f'(x)| \text{ subject to } f'(x) \text{ existing},
$$  
has maximum value $C = n < \infty$. (note: the slope of $f$ will reach the maximum as $x>\max(x_1,...,x_n)$ and the slope is equal to the number of $x_i$.)
$f$ is Lipschitz with Lipschitz constant $n$.

### part (g)
$f(x) = \frac{1}{2\sigma^2}(x-\mu)^2 + \lambda |x|$ for fixed values $\mu\in\mathbb{R}, \sigma > 0,$ and $\lambda > 0$ is convex.
  
Proof:  
We know that $f(x) = \frac{1}{2\sigma^2}(x-\mu)^2$ and $g(x)=|x|$ is convex.  
By the Theorem (Positive Weighted Sum of Convex is Convex),   
Let $a = 1, b = \lambda$, $a,b > 0$   
Therefore, $h:X\rightarrow\mathbb{R}$ defined by $h(x) = af(x) + bg(x) = \frac{1}{2\sigma^2}(x-\mu)^2 + \lambda |x|$ for all $x\in\mathbb{R}$ is convex on $\mathbb{R}$.

**Minimum value and minimizing solution**  
_As for minimum value and minimizing solution, we have to discuss by cases._
  
$f(x)$ =
  \begin{cases}
    \frac{1}{2\sigma^2}(x-\mu)^2 - \lambda x, & \text{for } x < 0 \\
    \frac{1}{2\sigma^2}(x-\mu)^2 + \lambda x, & \text{for } x \geq 0 
  \end{cases}  

As $x < 0$,  
$f'(x) = \frac{1}{\sigma^2}(x-\mu) - \lambda$  
Set $f'(x) == 0$, we get $x=\lambda\sigma^2+\mu$.  
since $x < 0$, then $\lambda\sigma^2+\mu < 0$, $\mu < -\lambda\sigma^2$.  
the minimum value is $f(\lambda\sigma^2+\mu)$.

As $x \geq 0$,  
$f'(x) = \frac{1}{\sigma^2}(x-\mu) + \lambda$  
Set $f'(x) == 0$, we get $x=-\lambda\sigma^2+\mu$.  
since $x > 0$, then $-\lambda\sigma^2+\mu > 0$, $\mu > \lambda\sigma^2$.  
the minimum value is $f(-\lambda\sigma^2+\mu)$.  

As $-\lambda\sigma^2<\mu<\lambda\sigma^2$, the minimizing solution is $0$ since the $f'(x)$ as $x<0$ is negative and the $f'(x)$ as $x>0$ is positive. The minimum value is $\frac{\mu^2}{2\sigma^2}$.

** 
Conclusion  
As $\mu < -\lambda\sigma^2$, minimizing solution: $x=\lambda\sigma^2+\mu$, minimum value: $-\frac{\lambda^2\sigma^2}{2}-\lambda\mu$  
As $\mu > \lambda\sigma^2$, minimizing solution: $x=-\lambda\sigma^2+\mu$, minimum value: $-\frac{\lambda^2\sigma^2}{2}+\lambda\mu$  
As $-\lambda\sigma^2<\mu<\lambda\sigma^2$, minimizing solution: $0$, minimum value: $\frac{\mu^2}{2\sigma^2}$.
